```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('Графики коллебаний для давления - Sheet1.csv')
```

```python
data.head()
```

```python
np.log(data['p1'].to_numpy())
```

```python
sns.set_style('darkgrid')

plt.figure(figsize=(10, 7))
plt.title('Улучшение p(t)')
plt.plot(data['t1'], np.log(data['p1'].to_numpy()), color = 'red', label = '1ый сет')
plt.plot(data['t2'], np.log(data['p2'].to_numpy()), color = 'green', label = '2ый сет')

k1, b1 = np.polyfit(data['t1'], np.log(data['p1'].to_numpy()), 1) 
k2, b2 = np.polyfit(data['t2'], np.log(data['p2'].to_numpy()), 1) 

plt.plot(data['t1'], data['t1'] * k1 + b1, color = 'red', label = '1ый сет МНК')
plt.plot(data['t2'], data['t2'] * k2 + b2, color = 'green', label = '2ый сет МНК')

plt.xlabel('Время t(c)')
plt.ylabel('Давление, мм.рт.стлб * 10^-4')
plt.legend()
#plt.plot(data['t3'], data['p3'].to_numpy())
#plt.plot(data['t4'], data['p4'].to_numpy())
```

```python
(k1 + k2) / 2
```

```python
(b1 + b2) / 2
```
